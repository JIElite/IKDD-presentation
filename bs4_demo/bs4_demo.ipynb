````python
import requests
from bs4 import BeautifulSoup


def fetchHTML(url):
    resp = requests.get(url)
    if resp.status_code != 200:
        return None
    
    return resp.text
````

````python
url = "https://www.ptt.cc/bbs/Mix_Match/index.html"
page_html = fetchHTML(url)
print(page_html)
````

Extract Title Name
==========

現在我們想要將 ptt 的標題都抓出來，我們應該怎麼去分析？

觀察一下可以看到 title name 都包裹在下面這樣的 html 裡面
```html
<div class="r-ent">
			<div class="nrec"><span class="hl f2">1</span></div>
			<div class="mark"></div>
			<div class="title">
```		
				<a href="/bbs/Mix_Match/M.1482250246.A.D5E.html">[女搭] 黑色單品的兩套穿搭</a>
```html			
			</div>
			<div class="meta">
				<div class="date">12/21</div>
				<div class="author">lin11325</div>
			</div>
		</div>

```

title name 會是 class="title" 下的 <a></a> 的內容

我們可以用 ** select ** 來存取這樣的元素

select method 的語法類似 **CSS** 存取上也十分方便


````python
soup = BeautifulSoup(page_html, "html.parser")
title_list = [title_html.text for title_html in soup.select(".title a")]
for title in title_list:
    print(title)
    
````
